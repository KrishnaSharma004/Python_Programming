```python
#value_counts : gives the frequency of the values
vk = pd.read_csv('/content/vk - Sheet1.csv', index_col = 'match').squeeze()
print(vk.value_counts())
```

```python
#sort_values : sorts the values  : not inplace (i.e., changes are not made in the original one)
print(vk.sort_values(ascending = False))
```

```python
#sort_index : sort the index ; not inplace
print(vk.sort_index(ascending = False))
```

```python
#for inplace changes we have a parameter : inplace (need to make it true)
vk_copy = vk.copy()
vk_copy.sort_values(inplace = True)
print(vk_copy)
```

```python
#slicing : last index is excluded
print(s[: 7]) # [start : stop : step]
```

```python
print(vk[::2])
```

```python
#drop duplicates  :
temp = pd.Series([1,1,1,3,3,3,5,5,5,5,6,6,6])
print(temp )
```

```python
temp.drop_duplicates()#by defalut it keeps only the first occurance
```

```python
#for keeping the last occurance :
temp.drop_duplicates(keep = 'last')
```

```python
#to check whether there are duplicates values and there positions
#and hoe many duplicates values in total are there
print(movies)
```

```python
movies.duplicated().sum()#how may duplicated movies are there
```

```python
temp2 = pd.Series([1,1,1,np.nan,2,2,4,5,np.nan,np.nan])
print(temp2)
```

```python
#isnull
print(temp2.isnull())
print("How many null values are there : ")
print(temp2.isnull().sum())
```

```python
#dropna :
print(temp2.dropna())
```

```python
#fillna :
print(temp2.fillna(temp2.median()))
```
